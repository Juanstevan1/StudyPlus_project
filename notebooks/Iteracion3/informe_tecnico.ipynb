## 3.  Modelo de Inteligencia MASINA

**Modelo Inteligente: Agente 'Pablo'**

Problema a Resolver: El objetivo del Agente 'Pablo' es personalizar los bloques de estudio en función de la carrera, materia a cursar y tema de interés del usuario, ajustando el enfoque académico según el perfil del estudiante.

**Mecanismo de Razonamiento:**

- **Fuente de Información:** Carrera del usuario, materia a cursar, y tema de interés.

- **Fuente de Activación:** El mecanismo se activa cuando el usuario proporciona información sobre su carrera, materia y tema.
Técnica de Inferencia: Utiliza lógica de predicados de primer orden para ajustar el contenido y enfoque del estudio basado en el perfil del usuario.

- **Estrategia de Razonamiento:** Deductiva. Ajusta el contenido de estudio a partir de la información proporcionada, diferenciando enfoques según la carrera y materia.

- **Lenguaje de Representación de Conocimiento:** OWL (Web Ontology Language), que permite estructurar y manipular el conocimiento en forma de ontologías.

- **Resultado Esperado:** El agente genera bloques de estudio personalizados y alineados con el perfil académico del usuario, diferenciando enfoques según la carrera y la materia.

**Marco Ontológico del Individuo:**

- **Nombre:** Perfiles académicos del usuario.
- **Descripción:** El sistema construye un perfil basado en la carrera, materia y tema de interés, ajustando el contenido del estudio según las especificaciones académicas.
- **Ontologías que lo integran:** Incluye ontologías relacionadas con carreras (ej., Administración, Ingeniería de Sistemas), materias (ej., Cálculo) y temas (ej., Derivadas).
- **Inferencias:** Deduce y ajusta el contenido de estudio según las diferencias de enfoque académico entre carreras y materias.

**Ontología de Dominio:**

- **Descripción:** Estructura el conocimiento sobre contenidos académicos y enfoques según diferentes perfiles.
- **Caracterización:** Define proposiciones y recomendaciones basadas en la carrera y materia del usuario, ajustadas a los temas de interés.
- **Fuente:** Utiliza una taxonomía global para categorizar temas académicos y enfoques de estudio.
- **Valores por Omisión:** El contenido y enfoque están asociados con la carrera del usuario (ej., Administración vs. Ingeniería de Sistemas).
- **Valores mín y máx:** Delimita el contenido de estudio y el enfoque según la carrera, materia y tema de interés del usuario.

**Funcionamiento del Modelo:**

- **Recepción de Datos:** El agente recibe datos sobre la carrera, materia y tema de interés del usuario.
- **Generación de Perfil:** Construye un perfil académico basado en la información recibida, comparándolo con ontologías preexistentes.
- **Inferencia y Adaptación:** Utiliza lógica de predicados para ajustar el contenido y enfoque del estudio según el perfil del usuario.
- **Recomendación Final:** Proporciona bloques de estudio personalizados, alineados con la carrera y materia del usuario, con un enfoque adaptado a su perfil académico.

**Ejemplo de Uso:**

- Usuario A (Carrera: Administración, Materia: Cálculo, Tema: Derivadas)
Recomendación: Contenido de derivadas adaptado a la perspectiva de Administración, con énfasis en aplicaciones prácticas y ejemplos relacionados con la gestión.

- Usuario B (Carrera: Ingeniería de Sistemas, Materia: Cálculo, Tema: Derivadas)
Recomendación: Contenido de derivadas con un enfoque técnico y matemático, incluyendo aplicaciones en programación y algoritmos.

## 6. Cronicas

# Diseño del Algoritmo
## Objetivo
Guiar al usuario en el proceso de estudio, generando bloques de texto relevantes basados en la carrera, materia y temática de interés, y recomendando el método de estudio más adecuado. El asistente ayudará al usuario a:
- Generar contenido de estudio basado en la carrera, materia y temática de interés.
- Recomendar un método de estudio que se ajuste al tema.
- Monitorear el progreso de los temas ya estudiados y sugerir temas no cubiertos.
- Sugerir temas adicionales relevantes para continuar el aprendizaje.

## Estructura Base del Conocimiento

### Crónica 1: Generación del contenido relevante del tema de estudio
**Descripción:** El asistente genera un bloque de contenido que explica el tema en función de la carrera, materia y tema proporcionados por el usuario.

```prolog
% Definición de la carrera, materia y tema
carrera(nombreCarrera).
materia(nombreMateria, nombreCarrera).
tema(nombreTema, nombreMateria).
contenido_tema(nombreTema, explicacion, metodo_estudio).

% Ejemplo
carrera(ingenieria).
materia(calculo, ingenieria).
tema(derivadas, calculo).
contenido_tema(derivadas, "Las derivadas representan la tasa de cambio instantánea de una función", ejercicios_practicos).
```

### Crónica 2: Recomendación de método de estudio
**Descripción** El asistente recomienda un método de estudio basado en el tema proporcionado por el usuario.

```prolog
% Recomendación de métodos de estudio
metodo_estudio(nombreTema, tipoMetodo).

% Ejemplo
metodo_estudio(derivadas, ejercicios_practicos).
metodo_estudio(integrales, diagramas).
```

### Crónica 3: Seguimiento del progreso del estudiante
**Descripción** El asistente lleva un registro de los temas que el estudiante ha cubierto y sugiere revisar o avanzar a nuevos temas si ya fueron estudiados.

```prolog
% Progreso del estudiante
tema_estudiado(agente, nombreTema).
siguiente_tema(nombreMateria, nombreTema).

% Ejemplo
tema_estudiado(agente1, derivadas).
siguiente_tema(calculo, integrales).
```

### Crónica 4: Sugerencia de temas no cubiertos
**Descripción** El asistente sugiere temas relevantes que aún no han sido estudiados para asegurar un avance completo en la materia.

```prolog
% Sugerencia de temas no cubiertos
tema_no_cubierto(agente, nombreTema).
siguiente_tema(nombreMateria, nombreTema).

% Ejemplo
tema_no_cubierto(agente1, integrales).
siguiente_tema(calculo, termodinámica).
```

# ¿Cómo trabaja cada crónica?

## Crónica 1: Generación del contenido relevante del tema de estudio

### Proceso:
1. El asistente recibe como entrada la carrera, la materia y el tema.
2. Busca en su base de conocimiento el contenido relevante para esa combinación.
3. Genera un bloque de estudio que explica el tema y proporciona un método de estudio.

### Ejemplo:
- **Entrada**: Carrera: "Ingeniería", Materia: "Cálculo", Tema: "Derivadas".
- **Salida**: "Las derivadas representan la tasa de cambio instantánea de una función. Método de estudio recomendado: ejercicios prácticos."

---

## Crónica 2: Recomendación de método de estudio

### Proceso:
1. Después de generar el contenido, el asistente recomienda un método de estudio basado en el tema.
2. Los métodos de estudio pueden incluir ejercicios, diagramas, lectura, etc.

### Ejemplo:
- **Entrada**: Tema: "Integrales".
- **Salida**: "Método de estudio recomendado: Usar diagramas para ilustrar áreas bajo la curva y resolver ejercicios prácticos."

---

## Crónica 3: Seguimiento del progreso del estudiante

### Proceso:
1. El asistente lleva un registro de los temas que el usuario ha estudiado.
2. Si un tema ya ha sido cubierto, el asistente sugiere revisarlo o avanzar a un nuevo tema.

### Ejemplo:
- **Entrada**: Tema: "Derivadas", ya estudiado.
- **Salida**: "Ya has estudiado 'Derivadas'. Te recomendamos revisar el tema de 'Integrales' como siguiente paso."

---

## Crónica 4: Sugerencia de temas no cubiertos

### Proceso:
1. El asistente sugiere nuevos temas que el estudiante aún no ha cubierto para avanzar en el aprendizaje.

### Ejemplo:
- **Entrada**: Carrera: "Ingeniería", Materia: "Física".
- **Salida**: "Te recomendamos estudiar 'Termodinámica' como siguiente tema relevante."

---

## Resumen
- **Crónica 1**: Generación de contenido de estudio. El asistente genera un bloque explicativo del tema proporcionado por el usuario.
- **Crónica 2**: Recomendación de método de estudio. Sugiere el método de estudio más adecuado para el tema basado en la carrera y materia.
- **Crónica 3**: Seguimiento del progreso del estudiante. Lleva un registro de los temas cubiertos y sugiere revisiones o nuevos temas si ya han sido estudiados.
- **Crónica 4**: Sugerencia de temas no cubiertos. Sugiere temas adicionales no cubiertos que el estudiante puede estudiar para avanzar.

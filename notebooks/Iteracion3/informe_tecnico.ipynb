## 1. Objetivo de la Iteración

El objetivo de esta iteración es equipar al Agente 'Pablo' con un modelo de conocimiento que le permita comprender y ajustarse a los diferentes enfoques académicos según mi carrera profesional. Este modelo incluye información sobre temas, carreras y asignaturas, facilitando que el agente genere bloques de estudio más alineados con los distintos perfiles.

## 4.  Base de Conocimiento en Lenguaje Natural

En lenguaje natural, el agente 'Pablo' funcionará con sentencias que ayudarán a ajustar recomendaciones académicas en función de la carrera, la materia y el tema de interés. El agente identificará similitudes y generará bloques de estudio específicos para cada perfil académico. 

**Sentencias:**

- A todos los que estudian Ingeniería de Sistemas y cursan Cálculo, les interesa aprender derivadas desde un enfoque técnico.

- A todos los que estudian Administración y cursan Cálculo, les interesa aprender derivadas aplicadas a negocios.

- A todos los que estudian Medicina y cursan Biología, les interesa aprender genética aplicada a la salud.

- A todos los que estudian Biotecnología y cursan Biología, les interesa aprender genética aplicada a la investigación científica.

- A todos los que estudian Psicología y cursan Estadística, les interesa aprender análisis de datos conductuales.

- A todos los que estudian Economía y cursan Estadística, les interesa aprender modelos predictivos para mercados.

- A todos los que estudian Derecho y cursan Filosofía, les interesa aprender teoría política aplicada.

- A todos los que estudian Sociología y cursan Filosofía, les interesa aprender filosofía del pensamiento social.

- A todos los que estudian Ingeniería Civil y cursan Física, les interesa aprender mecánica de estructuras.

- A todos los que estudian Física y cursan Mecánica, les interesa aprender dinámica aplicada a sistemas físicos.

- A todos los que estudian Diseño Gráfico y cursan Arte, les interesa aprender teoría del color.

- A todos los que estudian Historia del Arte y cursan Arte, les interesa aprender evolución de los movimientos artísticos.

- A todos los que estudian Ingeniería Eléctrica y cursan Cálculo, les interesa aprender derivadas aplicadas a sistemas eléctricos.

- A todos los que estudian Matemáticas Puras y cursan Cálculo, les interesa aprender derivadas desde un enfoque teórico.

- A todos los que estudian Comunicación y cursan Lenguaje, les interesa aprender sobre retórica aplicada a medios.

## 5. Base de Conocimiento en Lógica de Predicado

**Notación:**

- G(x, y, z) = A la persona x, que estudia la carrera y, le interesa aprender el tema z.

**Constantes:**

- Carreras: IngenieríaDeSistemas, Administración, Medicina, Biotecnología, Psicología, Economía, Derecho, Sociología, IngenieríaCivil, Física, DiseñoGráfico, HistoriaDelArte, IngenieríaEléctrica, MatemáticasPuras, Comunicación.
T
- Temas: Derivadas, Genética, AnálisisDeDatos, TeoríaPolítica, MecánicaEstructuras, TeoríaColor, Dinámica, Retórica.

- Enfoques: EnfoqueTecnico, EnfoqueNegocios, EnfoqueSalud, EnfoqueInvestigaciónCientífica, EnfoqueConductual, EnfoqueMercados, EnfoqueJurídico, EnfoquePensamientoSocial, EnfoqueConstrucción, EnfoqueSistemasFísicos, EnfoqueVisual, EnfoqueMovimientosArtísticos, EnfoqueSistemasEléctricos, EnfoqueTeórico, EnfoqueMedios.

**Variables:**

- x: Personas.
- y: Carreras (puede variar según lo que el usuario introduzca).
- z: Temas específicos que el usuario quiere aprender.

**Sentencias:**
